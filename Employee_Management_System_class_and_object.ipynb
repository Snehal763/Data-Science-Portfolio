```python
#Employee Management System
class Employee():
  empcount=0
  incmt=1.10
  def __init__(self,first,last,salary):
    self.first=first
    self.last=last
    self.salary=salary
    self.email=first.lower()+"."+last.lower()+"quastech.com"
    Employee.empcount +=1
  def fullname(self):
    return(f"{self.first.title()} {self.last.title()}")
  def increment(self):
    self.salary=int(self.salary * Employee.incmt)
```

```python
emp1=Employee("saniya","shaikh",25000)
emp2=Employee("snehal","shinde",22000)
emp3=Employee("soham","malgunkar",34000)
```

```python
class manager(Employee):
  def __init__(self,first,last,salary,employees=None):
    super().__init__(first,last,salary)
    if(employees is None):
      self.employees=[]
    else:
      self.employees=employees

  def addemp(self,emp):
    if emp not in self.employees:
      self.employees.append(emp)

  def remove(self,emp):
    if emp in self.employees:
      self.employees.remove(emp)

  def print_emps(self):
    for emp in self.employees:
      print("-->",emp.fullname())
```

```python
mgr1=manager("dhiraj","saroj",125000,[emp1,emp2,emp3])
mgr1.print_emps()
```
